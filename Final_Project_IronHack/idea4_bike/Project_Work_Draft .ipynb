## IMPORT AND CLEAN THE DATA

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the data using pandas 
df = pd.read_csv('hour.csv')
```

### Data Exploration

```python
#convert dates to datetime format 
df['dteday'] = pd.to_datetime(df['dteday'])
df['hr'] = pd.to_datetime(df['hr'], format='%H').dt.hour
```

```python
#drop month and year columns and extract from dteday column 
df.drop(['mnth', 'yr'], axis = 1, inplace = True) 
```

```python
#rename some columns for easy use 
df.rename(columns={'dteday': 'date','weathersit':'weather'},
          inplace=True, errors='raise')
```

## FEATURE ENGINEERING AND WRANGLING 

### Drop features that are not needed for model 

```python
# Drop some columns that will not serve our model 
df = df.drop(columns=['instant'])
# Also since we already have month year and day we will also drop the 'date' column 
df = df.drop(columns=['date'])

#df = df.drop(columns=['registered','casual'])
```

```python
#Convert numericals to categoricals where needed 
df[['weather', 'weekday', 'holiday','season', 'hr','month','year','workingday']] = df[['weather', 'weekday', 'holiday','season', 'hr','month','year','workingday']].apply(lambda x: x.astype('category'))
```

```python
df.info()
```

```python
# seperate categorical and numerical columns 
```

```python
df_cat =df.select_dtypes(include=['category'])
```

```python
df_num = df._get_numeric_data()
```

```python
# get dummies for categorical data 
df_cat = pd.get_dummies(df_cat, drop_first=True)
df_cat
```

### Drop target column for prediction 

```python
X=df_num.drop(['cnt'],axis=1)
y=df['cnt']
y
```

```python
#scale the numeric data 
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X)
x_scaled = scaler.transform(X)
X = pd.DataFrame(x_scaled, columns = X.columns)
```

```python
X_full = pd.concat([X, df_cat], axis=1)
```

```python
X_full.head(1)
```

### Split into test and train before running the model 

```python
#Split data into Test and train sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 60)
```

```python
y_test
```

## BUILD MODELS AND TEST FOR ACCURACY

### First Model - Linear Regression 

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
linearRegressor = LinearRegression()
linearRegressor.fit(X_train, y_train)
y_pred = linearRegressor.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test,y_pred)
print("Mean Squared Error:",mse)
print("R score:",r)
print("Mean Absolute Error:",mae)
```

```python
len(y_pred)
```

### Second Model - Decision Tree 

```python
# Decision Tree
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 30)
regressor.fit(X_train, y_train)
y_pred_dt = regressor.predict(X_test)
y_pred_dt
```

```python
#Test for Model Accuracy 
mse = mean_squared_error(y_test, y_pred_dt)
r = r2_score(y_test, y_pred_dt)
mae = mean_absolute_error(y_test,y_pred_dt)
print("Mean Squared Error:",mse)
print("R score:",r)
print("Mean Absolute Error:",mae)
```

### Third Model - Random Forest 

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train);
y_pred_rf = rf.predict(X_test)
y_pred_rf
```

```python
#Test for Model Accuracy 
mse = mean_squared_error(y_test, y_pred_rf)
r = r2_score(y_test, y_pred_rf)
mae = mean_absolute_error(y_test,y_pred_rf)
print("Mean Squared Error:",mse)
print("R score:",r)
print("Mean Absolute Error:",mae)
```

## FEATURE ENGINEERING part 3 

### drop outliers in windspeed 

```python
X_full.boxplot(column=['windspeed'])
```

```python
#merge target dataframe 
X_full_2 = pd.concat([X_full, y], axis=1)
X_full_2.head(1)
```

```python
#drop outliers in windspeed 
Q1 = X_full_2['windspeed'].quantile(0.25)
Q3 = X_full_2['windspeed'].quantile(0.75)
IQR = Q3 - Q1

X_full_2 = X_full_2[~((X_full_2['windspeed'] < (Q1 - 1.5 * IQR)) |(X_full_2['windspeed'] > (Q3 + 1.5 * IQR)))]
```

```python
X_full_2.boxplot(column=['windspeed'])
```

```python
X_full_2 = X_full_2.reset_index(drop=True)
```

```python
X_full_2.info()
```

```python
#drop target and index and assign target to y 
X=X_full_2.drop(['cnt','registered','casual'],axis=1)
y= X_full_2['cnt']
```

```python
#split data into train and Test model 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 30)
```

## BUILD FINAL MODEL 

```python
regressor = DecisionTreeRegressor(random_state = 30)
regressor.fit(X_train, y_train)
y_pred_dt3 = regressor.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred_dt3)
r = r2_score(y_test, y_pred_dt3)
mae = mean_absolute_error(y_test,y_pred_dt3)
print("Mean Squared Error:",mse)
print("R score:",r)
print("Mean Absolute Error:",mae)
```
