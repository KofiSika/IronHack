```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://freemeteo.es/eltiempo/barcelona/7dias/lista/?gid=3128760&language=spanish&country=spain"
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content,"html.parser")
```

```python
soup
```

```python
all = []

weather = soup.find_all("div",{"class":"day"})

for i in range(len(weather)):
    all.append(weather[i].text)
    
```

```python
all = [x.split("\n") for x in all]
```

```python
all
```

```python
date = [x[1] for x in all]
```

```python
date
```

```python
temperature = [x[3] for x in all]
```

```python
temperature
```

```python
wind = [x[5] for x in all]
```

```python
wind
```

```python
precipitation = [x[6] for x in all]
```

```python
weather = [x[11] for x in all]
```

```python
weather
```

```python
weather_df = pd.DataFrame({"date":date, "temperature":temperature, "wind":wind,"weather":weather, "precipitation":precipitation})
```
